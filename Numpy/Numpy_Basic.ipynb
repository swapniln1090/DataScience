## Numpy vs List

```python
# list

x = list(range(10000000))
y = list(range(10000000,20000000))

start_time = time.time()

res = [x+y for x,y in zip(x,y)]

print(time.time()-start_time)
```

```python
res
```

```python
x = np.arange(1000000)
y = np.arange(1000000,2000000)

start_time = time.time()

res1 = np.add(x,y)

print(time.time()-start_time)
```

```python
res1
```

## Reshape

```python
arr11 = np.arange(24).reshape(2,3,4) # reshape can only happen if no of given array elements is equal to the output of multiplication of reshape size
#i.e. here total ele is 24 and 2*3*4 = 24 . hen
arr11
```

```python
arr11 =  np.arange(24).reshape(3,3,4) #the product of 3*3*4 does not match element nos.
```
