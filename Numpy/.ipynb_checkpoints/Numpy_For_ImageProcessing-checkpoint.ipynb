## Image Statastics

```python
img
```

```python
nrows, ncols, channels = img.shape
print(nrows, ncols, channels)
```

```python
row, col = np.ogrid[:nrows, :ncols]
```

```python
col
```

```python
cnt_row, cnt_col = (nrows/2) , (ncols/2)
```

```python
outer_disk_mask = ((row - cnt_row)**2 + (col - cnt_col)**2 > cnt_row)
outer_disk_mask
```

```python
outer_disk_mask[200,129] #somewhere in middle
```

```python
new_img = img.copy()
new_img[outer_disk_mask] = 0
```

```python
plt.imshow(new_img)
```

## Image Chanels

```python
img.shape
```

```python
r = img[: , : , 0]
g = img[: , : , 1]
b = img[: , : , 2]
```

```python
r.shape
```

```python
opt = [img, r, g, b]
titles = ['Original', 'Red' , 'Green', 'Blue']
my_cmap = ['dummy','Reds', 'Greens', 'Blues'] #dummy becz we want original to be displayed as is. hence in below code its already written that way
```

```python
for i in range(len(opt)):
    plt.subplot(2, 2, i+1)
    
    plt.axis('off')
    plt.title(titles[i])
    if i == 0:
        plt.imshow(opt[i])
    else:
        plt.imshow(opt[i], cmap = 'gray')
plt.show()
```

```python
for i in range(len(opt)):
    plt.subplot(2, 2, i+1)
    
    plt.axis('off')
    plt.title(titles[i])
    if i == 0:
        plt.imshow(opt[i])
    else:
        plt.imshow(opt[i], cmap = my_cmap[i])
plt.show()
```

```python
# if we just have r, g,b separate image then we can create a original image by stacking on each other 
output = np.dstack((r, g, b))
plt.imshow(output)
```
