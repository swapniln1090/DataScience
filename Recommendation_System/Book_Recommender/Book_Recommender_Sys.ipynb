# Collarative Filtering

```python
books_with_rating
```

#### Let's make a matrix of UserIds and Books w.r.t their Ratings.
For this lets -
1. get all the users who have made ratings for diffrent books which is more than 200. (We can choose best no which gives us best results)
2. Get all books who has ratings more than 50 times

###### W.R.T User-ID

```python
#
x = books_with_rating.groupby("User-ID").count()["Book-Rating"] >200
x
```

```python
imp_user = x[x].index
imp_user
```

```python
books_with_rating[books_with_rating["User-ID"].isin(imp_user)]
```

```python
filtered_rating = books_with_rating[books_with_rating["User-ID"].isin(imp_user)]
```

##### 2. Out of this filtered DF, lets get all Books-titles which has count of ratings greater than 50

```python
filtered_rating.groupby("Book-Title").count()["Book-Rating"]
```

```python
filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
```

```python
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
famous_books = y[y].index
famous_books
```

```python
filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
final_ratings = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

#### Create pivot table

```python
piv_tab = final_ratings.pivot_table(index="Book-Title", columns='User-ID', values = 'Book-Rating')
piv_tab
```

```python
piv_tab.fillna(0, inplace=True)
```

```python
piv_tab
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(piv_tab)
similarity_scores
```

```python
len(similarity_scores[0]) , similarity_scores.shape
```

```python
# lets sort it 
sorted(list(enumerate(similarity_scores[0])), key= lambda x:x[1], reverse=True) # becuase without key it takes 1st position of every element. 
                                                                    #We have to sort according to our similaity score and not with enumerated sequence.
                                                                    
```

```python
# Create a function which when given a Book, should give most similar 5 books as recomendation
```

```python
def recommender(book_name):
    
    index = np.where(piv_tab.index==book_name)[0][0]
    book_score = similarity_scores[index]
    similar_items = sorted(list(enumerate(book_score)),key = lambda x:x[1], reverse=True)[1:6] # Let's take only 5 item. Ist one is always the same book so startinging from index 1

    # now lets get all similar items
    for i in similar_items:
        print(piv_tab.index[i[0]]) # Because we need index to 
    
    #return suggestions
```
